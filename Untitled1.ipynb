```python
from qiskit import QuantumCircuit
from qiskit import IBMQ
from qiskit import transpile
import numpy as np
from qiskit.tools.monitor import job_monitor, backend_monitor, backend_overview

from qiskit import Aer
sv_sim = Aer.get_backend('aer_simulator')

IBMQ.load_account()
provider = IBMQ.get_provider('ibm-q')
quito = provider.get_backend('ibmq_quito')


initialize = QuantumCircuit(2, 2)

#X gate on qubit 0
initialize.x(0)
#X gate on qubit 1
initialize.x(1)
#Hadmard gate on 
initialize.h(0)
#cnot gate on 0 and 1
initialize.cx(0, 1)
#draw the circuit
initialize.draw('mpl')

rotation_circuits = []

rotatexy45 = QuantumCircuit(2, 2)
rotatexy45.rz(np.pi/4, 1)
rotation_circuits.append(rotatexy45)


rotatexz45 = QuantumCircuit(2, 2)
rotatexz45.rz(np.pi/2, 1)
rotation_circuits.append(rotatexz45)

rotateyz90 = QuantumCircuit(2, 2)
rotateyz90.rz(np.pi/4, 0)
rotateyz90.rz(np.pi/2, 1)
rotation_circuits.append(rotateyz90)

rotatexy60 = QuantumCircuit(2, 2)
rotatexy60.rz(np.pi/3, 1)
rotation_circuits.append(rotatexy60)

rotatexz60 = QuantumCircuit(2, 2)
rotatexz60.rz(2*np.pi/3, 1)
rotation_circuits.append(rotatexz60)

rotateyz120 = QuantumCircuit(2, 2)
rotateyz120.rz(np.pi/3, 0)
rotateyz120.rz(2*np.pi/3, 1)
rotation_circuits.append(rotateyz120)

rotation_circuits[0].draw()

meas = QuantumCircuit(2, 2)
meas.barrier(range(2))
#apply hadmard gates to both qubits
meas.h((0, 1))
# add barrier
meas.barrier(range(2))
#add z measurement
meas.measure(range(2), range(2))
meas.draw()

# The Qiskit circuit object supports composition using
# the addition operator.
# print(meas.cregs[0])
# qc = circ.compose(meas)

#drawing the circuit
#qc.draw()

#build circuits
quantum_circuits = []
for rot in rotation_circuits:
#     circuit = QuantumCircuit(2, 2)
#     circuit.compose(initialize)
#     circuit.compose(rot)
#     circuit.compose(meas)
    circuit = initialize + rot + meas
    quantum_circuits.append(circuit)
    circuit.draw()
    

quantum_circuits[1].draw()

# Use Aer's qasm_simulator

# Execute the circuit on the qasm simulator.
# We've set the number of repeats of the circuit
# to be 1024, which is the default.
job_ideal = sv_sim.run(transpile(quantum_circuits, sv_sim), shots=10000)
job_real = quito.run(transpile(quantum_circuits, quito), shots=10000)
job_monitor(job_real)

# Grab the results from the job.
result_sim = job_sim.result()
result_real = job_real.result()

counts = result_sim.get_counts()
print(counts)
from qiskit.visualization import plot_histogram
plot_histogram(counts[0])

import qiskit.tools.jupyter
%qiskit_version_table


```
